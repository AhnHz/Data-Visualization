```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['font.family']='d2coding'
plt.rcParams['axes.unicode_minus']=False
```

# 샘플데이터

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

# 요일별 팁 평균

## matplotlib으로 시각화

### 데이터 가공
* 요일별 팁 평균 계산
* 그룹핑 : 데이터프레임.groupby(그룹기준컬럼)[통계적용컬럼].통계함수

```python
tipm = tips.groupby('day')['tip'].mean()
```

### 데이터 시각화
* plt.bar(x,y)

```python
plt.bar(tipm.index, tipm, width=0.7, color='darkgray')
plt.title('요일별 팁 평균', size=15)
plt.show()
```

## seaborn으로 그리기
* <b>`sns.barplot(data=데이터프레임명, x=x축컬럼, y=y축컬럼)`</b>
* x축데이터로 그룹핑한 y축데이터의 평균값을 계산하여 그래프를 그려준다.
* 신뢰구간(CI:Confidence Interval)을 함께 표시<br>

```python
sns.barplot(data=tips, x='day', y ='tip', ci=None, width=0.7)
plt.show()
```

# 요일별 팁 합계

## matplotlib으로 그리기

```python
tipsum = tips.groupby('day')['tip'].sum()
tipsum
```

```python
plt.bar(tipsum.index, tipsum, width=0.7, color='lightblue')
plt.title('요일별 팁 합계', size=15)
plt.show()
```

## seaborn으로 그리기
* estimator = 통계함수

```python
sns.barplot(data=tips, x='day', y='tip', errorbar=None, estimator=sum)
plt.title('요일별 팁 합계', size=15)
plt.show()
```

# 요일별 팁 합계를 흡연여부로 비교
* hue : y를 그룹핑할 컬럼

```python
sns.barplot(data=tips, x='day', y='tip', errorbar=None, estimator=sum, hue='smoker')
plt.title('요일별 팁 합계 (흡연 여부 비교)', size=15)
plt.show()
```

* hue 색상 변경 : palette = 구분:색상 딕셔너리

```python
sns.barplot(data=tips, x='day', y='tip', errorbar=None, estimator=sum, hue='smoker',
           palette={'Yes':'lightgray', 'No':'skyblue'})

plt.title('요일별 팁 합계 (흡연 여부 비교)', size=15)
plt.show()
```

* pyplot의 메소드로 그리드 추가

```python
sns.barplot(data=tips, x='day', y='tip', errorbar=None, estimator=sum, hue='smoker',
           palette={'Yes':'lightgray', 'No':'skyblue'})

plt.grid(ls=':')
plt.title('요일별 팁 합계 (흡연 여부 비교)', size=15)
plt.show()
```
