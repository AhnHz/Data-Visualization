그래프의 특정 위치에 설명을 추가할 수 있다.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 그래프에 한글폰트 설정
plt.rcParams['font.family']='d2coding'

# 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False
```

# 텍스트 추가하기
* plt.text(x좌표, y좌표, 텍스트)
* rotation=회전각도
* ha : horizontal alignment
* va : vertical alignment
* 텍스트 상자<br>
`bbox = {'boxstyle':상자스타일, 'fc':facecolor,'ec':edgecolor,...}`<br>
boxstyle : 'round'/'square'

```python
plt.plot([1,2,3,4], 'ko')
plt.text(2.1,3, '(x:2, y:3)', ha='left', rotation=30, fontsize=12,
        bbox={'boxstyle':'round', 'fc':'lightgray', 'ec':'gray', 'alpha':0.5})

plt.axhline(3, ls=':', alpha=0.5, lw=1)
plt.axvline(2, ls=':', alpha=0.5, lw=1)
plt.plot(2,3, 'bo')
plt.show()
```

# 화살표와 텍스트 추가하기
* `plt.annotate('텍스트',xy=(화살표x,화살표y), xytext=(텍스트x,텍스트y), arrowprops=화살표속성(딕셔너리))`<br>
<br>
* 화살표 속성<br>
`width`        The width of the arrow in points<br>
`headwidth`    The width of the base of the arrow head in points<br>
`headlength`   The length of the arrow head in points<br>
`shrink`       Fraction of total length to shrink from both ends<br>

```python
plt.plot([1,2,3,4], 'ko')
plt.annotate('(x:1, y:2)', xy=(1,2), xytext=(1.25, 2.25), fontsize=12,
             arrowprops={'width':2, 'headwidth':10, 'headlength':12, 'shrink':2, 'fc':'lightgray'})

plt.axhline(2, color='lightpink', ls=':', alpha=0.7)
plt.axvline(1, color='lightpink', ls=':', alpha=0.7)
plt.plot(1,2, 'ro')

plt.show()
```
