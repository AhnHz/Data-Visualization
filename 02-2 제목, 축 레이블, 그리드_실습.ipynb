## 샘플 데이터 
* 월별 몸무게 변화<br>
어떤 사람이 체중 관리를 위해 월별 몸무게를 기록하고 있음. <br>
월별 몸무게 변화를 시각화하고자 함<br>
* 1월:80kg, 2월:78kg, 3월:75kg, 4월:73kg, 5월:70kg

# 데이터프레임 만들기

```python
import pandas as pd
```

```python
df = pd.DataFrame({'월':[1,2,3,4,5], '몸무게':[80,78,75,73,70]})
df
```

## 그래프 그리기

```python
x = df['월']
y = df['몸무게']

plt.plot(x,y,'ro-', mec='b', mfc='b', lw=2, ms=5)
plt.show()
```

# 제목
* plt.title('제목')
* loc : 제목 위치('left','center','right')
* pad : 타이틀과 그래프와의 간격
* color : 폰트색상
* fontsize : 폰트사이즈
* fontweight : 폰트굵기('normal','bold','heavy','light','ultrabold','ultralight')
* fontfamily : 폰트

```python
plt.plot(x,y,'ro-', mec='b', mfc='b', lw=2, ms=5)
plt.title('월별 몸무게 변화', loc='center', pad=10, color='black', fontsize=15, fontweight='bold'
          , fontfamily='D2Coding')
plt.show()
```

# x축, y축 레이블
* plt.xlabel('레이블')     
* plt.ylabel('레이블')
* loc : 위치('left','center','right' / 'bottom', 'center','top')
* labelpad : 레이블과 그래프와의 간격
* color : 폰트색상
* fontsize : 폰트사이즈
* fontfamily : 폰트

```python
plt.plot(x,y,'ro-', mec='b', mfc='b', lw=2, ms=5)
plt.title('월별 몸무게 변화', loc='center', pad=10, color='black', fontsize=15, fontweight='bold'
          , fontfamily='D2Coding')
plt.xlabel('월',loc='right',labelpad=10, color='k',fontsize=10)
plt.ylabel('몸무게(kg)', loc='top',labelpad=10, color='k', fontsize=10)
plt.show()
```

# 그리드
* plt.grid(True)
* axis : 그리드 방향
* ls, lw, color, alpha 등의 속성을 지정할 수 있음

## 기본 그리드
* plt.grid()

```python
plt.plot(x,y,'ro-', mec='b', mfc='b', lw=2, ms=5)
plt.title('월별 몸무게 변화', loc='center', pad=10, color='black', fontsize=15, fontweight='bold'
          , fontfamily='D2Coding')
plt.xlabel('월',loc='right',labelpad=10, color='k',fontsize=10)
plt.ylabel('몸무게(kg)', loc='top',labelpad=10, color='k', fontsize=10)

plt.grid()
plt.show()
```

## x축 그리드

```python
plt.plot(x,y,'ro-', mec='b', mfc='b', lw=2, ms=5)
plt.title('월별 몸무게 변화', loc='center', pad=10, color='black', fontsize=15, fontweight='bold'
          , fontfamily='D2Coding')
plt.xlabel('월',loc='right',labelpad=10, color='k',fontsize=10)
plt.ylabel('몸무게(kg)', loc='top',labelpad=10, color='k', fontsize=10)

plt.grid(axis='x')
plt.show()
```

## y축 그리드

```python
plt.plot(x,y,'ro-', mec='b', mfc='b', lw=2, ms=5)
plt.title('월별 몸무게 변화', loc='center', pad=10, color='black', fontsize=15, fontweight='bold'
          , fontfamily='D2Coding')
plt.xlabel('월',loc='right',labelpad=10, color='k',fontsize=10)
plt.ylabel('몸무게(kg)', loc='top',labelpad=10, color='k', fontsize=10)

plt.grid(axis='y')
plt.show()
```

## 그리드 스타일링

```python
plt.plot(x,y,'ro-', mec='b', mfc='b', lw=2, ms=5)
plt.title('월별 몸무게 변화', loc='center', pad=10, color='black', fontsize=15, fontweight='bold'
          , fontfamily='D2Coding')
plt.xlabel('월',loc='right',labelpad=10, color='k',fontsize=10)
plt.ylabel('몸무게(kg)', loc='top',labelpad=10, color='k', fontsize=10)

plt.grid(color='skyblue', alpha=0.5, ls='--')
plt.show()
```
