```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 그래프에 한글폰트 설정
plt.rcParams['font.family']='d2coding'

# 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 그리드 스타일
plt.rcParams['grid.linestyle'] = '--'
plt.rcParams['grid.alpha'] = 0.3
```

# 수평선, 수직선 그리기

## 수평선 그리기
* `plt.axhline(y좌표, x축시작위치, x축끝위치)`<br>
수평선의 길이가 1이라고 했을 때 x축시작위치, x축끝위치를 지정한다.<br>
따로 지정하지 않으면 x축 전범위에 걸쳐 그려진디ㅏ.
* `plt.hlines(y, x축시작좌표, x축끝좌표)`<br>

```python
plt.plot([1,2,3,4], 'ko')

plt.axhline(4, 0.1, 0.9, color='lightgray', ls='--', linewidth=2)  # 수평선 전체 길이가 1
plt.hlines(2, 2, 7, color='lightblue', linewidth=2)

plt.show()
```

## 수직선 그리기
* `axvline(x좌표, y축시작위치, y축끝위치)`<br>
수직선의 길이가 1이라고 했을 때 y축시작위치, y축끝위치를 지정한다.<br>
따로 지정하지 않으면 y축 전범위에 걸쳐 그려진디ㅏ.
* `vlines(x, y축시작좌표, y축끝좌표)`<br>

```python
plt.plot([1,2,3,4], 'ko')

plt.axvline(1.5, c='lightgray', linewidth=2)
plt.vlines(1, 1.5, 3.5, color='lightblue', linewidth=2, ls=':')
plt.show()
```

# 예제 : 요일별 판매 테이블 수

## 데이터

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head(3)
```

## 요일 별 테이블 수
* 요일별 데이터 수

```python
day = tips['day'].value_counts()
day
```

```python
plt.figure(figsize=(15,3))

plt.barh(day.index, day.values)
plt.xticks(range(0,91,1), rotation=90)

plt.grid(axis='x', alpha=0.5)
plt.axvline(day['Fri'], color='k')
plt.axvline(day['Thur'], color='k')
plt.axvline(day['Sun'], color='k')
plt.axvline(day['Sat'], color='k')

plt.show()
```
