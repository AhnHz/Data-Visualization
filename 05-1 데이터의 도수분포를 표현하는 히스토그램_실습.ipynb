# 히스토그램
* 자료의 도수분포를 나타내는 그래프<br>
수치를 나타내는 자료를 일정 구간(계급)으로 나누어 각 구간별 값의 개수(도수)를 나타낸다.<br>

* x축은 계급(값의 구간), y축은 도수(구간별 값의 개수)를 나타냄

## 샘플데이터

## 히스토그램

### 기본 히스토그램
* <b>`plt.hist(data)`</b><br>
default : 10개의 구간으로 나누어 도수분포를 보여준다.

```python
plt.hist(scores)
plt.show()
```

### 구간의 개수 변경 
<b>` bins=구간의 개수 `</b>

```python
plt.hist(scores, bins=20)
plt.show()
```

### 누적 히스토그램
<b>`cumulative=True`</b>

```python
plt.hist(scores, bins=20, cumulative=True)
plt.show()
```

### 범위 지정
<b>`range=(min,max)`</b>

```python
plt.hist(scores, bins=20, cumulative=True, range=(50,100))
plt.show()
```

### 밀도 표시
<b>`density=True`</b><br>
도수를 총 개수로 나눈 수치를 y축에 표시

```python
plt.hist(scores, bins=20, density=True)
plt.show()
```

```python
plt.hist(scores, bins=20, density=True, cumulative=True)
plt.show()
```

### 가로 히스토그램
<b>`orientation='horizontal'`</b>

```python
plt.hist(scores, bins=20, orientation='horizontal')
plt.show()
```

### 막대 스타일

* 선만 표시 <b>`histtype='step'`</b>

```python
plt.hist(scores, bins=20, histtype='step')
plt.show()
```

`rwidth=막대폭(0~1)` `color=막대색` `alpha=투명도`<br>
`edgecolor(ec)=선색` `linewidth(lw)=선두께` `linestyle(ls)=선스타일`<br>
`hatch=패턴`



```python
plt.hist(scores, bins=20, color='lightgray', alpha=0.5,
        edgecolor='gray', linewidth=2, ls='--', hatch='/')

plt.grid(axis='y', ls=':')

plt.title('학생 125명의 점수', size=15)
plt.xlabel('점수')
plt.ylabel('인원')
plt.show()
```
