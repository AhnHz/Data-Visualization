# 샘플데이터

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.tail()
```

```python
# 연도별 데이터 수
flights['year'].value_counts()
```

```python
# 월별 데이터 수
flights['month'].value_counts()
```

# 연도별 승객수의 변화

## matplotlib으로 시각화
* plt.plot(x,y)

### 데이터 가공

```python
f_year = flights.groupby('year')['passengers'].sum()
f_year
```

### 시각화

```python
plt.plot(f_year, 'ko--')
plt.xlabel('year')
plt.ylabel('passengers')
plt.show()
```

## seaborn
* `sns.lineplot(data=데이터프레임, x=x축컬럼, y=y축컬럼, estimator=통계함수)`<br>
estimator를 생략하면 평균으로 통계를 적용함

* 전체 데이터로 차트를 그리면 신뢰구간 표시

```python
sns.lineplot(data=flights, x='year', y='passengers', errorbar=None, estimator=sum, color='k', ls='--', marker='o')
plt.show()
```

# 연도-월별 승객수의 변화

## matplotlib으로 시각화

### 데이터 가공

```python
f_pivot = flights.pivot(index='year', columns='month', values='passengers')
f_pivot
```

### 시각화

```python
for i in range(len(f_pivot)):
    plt.plot(f_pivot.iloc[i], label=f_pivot.index[i])
    
plt.legend()
plt.show()
```

## seaborn으로 시각화

```python
sns.lineplot(data=flights, x='month', y='passengers', errorbar=None, estimator=sum, hue='year')
plt.show()
```
