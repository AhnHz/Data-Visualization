# 라이브러리 임포트

```python
import matplotlib.pyplot as plt
```

# 한글폰트 사용

```python
# 폰트 확인하기
import matplotlib.font_manager as fm
[f.name for f in fm.fontManager.ttflist if 'D2Coding' in f.name]
```

```python
# 한글폰트 지정
plt.rcParams['font.family'] = 'D2Coding'

# 한글폰트 사용 시 -기호 깨지는 문제 해결
```

```python
# 그래프로 확인
x = ['봄','여름','가을','겨울']
y = [19,30,15,-10]

# 그래프 그리기
plt.plot(x,y)
plt.show()
```

# 색상, 마커, 선

## 색상
* color=색상
* 색상 이름을 사용한다.
https://matplotlib.org/2.0.2/examples/color/named_colors.html
* 자주 사용되는 색깔은 약자를 사용할 수 있다.(blue:b, green:g, red:r, cyan:c, magenta:m, yellow:y, black:k, white:w)
* hex code를 사용할 수 있다.

```python
x = ['봄','여름','가을','겨울']
y = [19,30,15,-10]
```

```python
# 색상이름

plt.plot(x, y, color='cyan')
plt.show()
```

```python
# 색상이름 약자(blue:b, green:g, red:r, cyan:c, magenta:m, yellow:y, black:k, white:w)

plt.plot(x, y, color='g')
plt.show()
```

```python
# 색상코드

plt.plot(x, y, color='#708090')
plt.show()
```

## 마커
* marker : 마커종류 (* . , o v ^ < > 1 2 3 4 s p * h H + x D d)
* markersize, ms : 마커사이즈
* markeredgecolor, mec : 마커 선 색깔
* markeredgewidth, mew : 마커 선 굵기
* markerfacecolor, mfc : 마커 내부 색깔

```python
plt.plot(x, y, marker='D')
plt.show()
```

## 선
* linestyle, ls : 선스타일     
``'-'``          solid line style    
``'--'``         dashed line style    
``'-.'``         dash-dot line style        
``':'``          dotted line style    

* linewidth, lw : 선 굵기

```python
plt.plot(x, y, ls='-.')
plt.show()
```

## [색상][마커][선]
* [color][marker][line]

### 색상

```python
plt.plot(x, y, color='r')
plt.show()
```

### 마커

```python
plt.plot(x, y, marker='o')
plt.show()
```

### 선

```python
plt.plot(x, y, ls='--')
plt.show()
```

### 색상,마커

```python
plt.plot(x, y, color='g', marker='^')
plt.show()
```

### 색상,선

```python
plt.plot(x, y, color='b', ls=':')
plt.show()
```

### 마커, 선

```python
plt.plot(x, y, marker='d', ls='-')
plt.show()
```

### 색상,마커,선

```python
plt.plot(x, y, color='Teal', marker='*', ls='-')
plt.show()
```
